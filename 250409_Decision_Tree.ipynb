# 결정 트리

## 로지스틱 회귀로 와인 분류하기

- 화이트와인 레드와인 분류

```python
import pandas as pd
# 와인의 데이터 셋 불러오기
wine = pd.read_csv('https://bit.ly/wine-date')
```

```python
wine.head()
```

```python
wine.info() # 데이터 확인하는 코드
```

```python
wine.describe() # 데이터 정보를 요약해서 확인하기 쉬움
```

```python
data = wine[['alcohol', 'sugar', 'pH']].to_numpy()
target = wine['class'].to_numpy()
```

```python
from sklearn.model_selection import train_test_split
# 실제 훈련 데이터셋과 테스트 데이터셋을 분류
train_input, test_input, train_target, test_target = train_test_split(
    data, target, test_size=0.2, random_state=42)
```

```python
print(train_input.shape, test_input.shape) # 8:2로 분류
```

```python
from sklearn.preprocessing import StandardScaler
# 표준화 작업 코드
ss = StandardScaler()
ss.fit(train_input)

train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(train_scaled, train_target)
# 정확도 계산
print(lr.score(train_scaled, train_target))
print(lr.score(test_scaled, test_target))
```

### 설명하기 쉬운 모델과 어려운 모델

```python
print(lr.coef_, lr.intercept_) # 기울기와 편향
```

## 결정 트리

```python
from sklearn.tree import DecisionTreeClassifier
# 결정트리 정의
dt = DecisionTreeClassifier(random_state=42)
dt.fit(train_scaled, train_target) # 학습 시키기
# 스코어 값
print(dt.score(train_scaled, train_target))
print(dt.score(test_scaled, test_target))
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
# 결정트리 그래프 그려보기
plt.figure(figsize=(10,7))
plot_tree(dt)
plt.show()
```

```python
plt.figure(figsize=(10,7))
plot_tree(dt, max_depth=1, filled=True, feature_names=['alcohol', 'sugar', 'pH']) # 그래프로 확인하기 위해서 1개만 보게함
plt.show()
```

### 가지치기

```python
dt = DecisionTreeClassifier(max_depth=3, random_state=42) # 과적합 방지
dt.fit(train_scaled, train_target)

print(dt.score(train_scaled, train_target))
print(dt.score(test_scaled, test_target))
```

```python
plt.figure(figsize=(20,15))
plot_tree(dt, filled=True, feature_names=['alcohol', 'sugar', 'pH'])
plt.show()
```

```python
dt = DecisionTreeClassifier(max_depth=3, random_state=42) # 과적합 방지
dt.fit(train_input, train_target)

print(dt.score(train_input, train_target))
print(dt.score(test_input, test_target))
```

```python
plt.figure(figsize=(20,15))
plot_tree(dt, filled=True, feature_names=['alcohol', 'sugar', 'pH'])
plt.show()
```

```python
print(dt.feature_importances_) # 당도 특성의 중요도가 제일 높음
```

## 확인문제

- 트리의 좌우 노드가 균등하지 않다

```python
# 최소 불순도(min_impurity_decrease) 를 0.0005로 설정
dt = DecisionTreeClassifier(min_impurity_decrease=0.0005, random_state=42)
dt.fit(train_input, train_target)

print(dt.score(train_input, train_target))
print(dt.score(test_input, test_target))
```

```python
plt.figure(figsize=(20,15), dpi=300)
plot_tree(dt, filled=True, feature_names=['alcohol', 'sugar', 'pH'])
plt.show()
```
